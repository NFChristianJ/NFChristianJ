```python
# Importation des bibliotheques necessaires

from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# URL de la page Wikipédia dont nous allons reccuperer les données

url = "https://fr.wikipedia.org/wiki/Liste_des_plus_grandes_entreprises_par_chiffre_d%27affaires"

# Faire la requête HTTP
response = requests.get(url)
```

```python
# Verifions si la requete a reussi.

response.status_code
```

```python
# Créons d'un objet BeautifulSoup

soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Trouvons toutes les tables avec la classe wikitable

tables = soup.find_all("table", {"class": "wikitable"})
```

```python
# Vérifions si des tables ont été trouvées
if tables:
    print(f"Nombre de tables trouvées : {len(tables)}")
else:
    print("Aucune table trouvé")
```

```python
# Affichons un aperçu de chaque table pour identifier la bonne

for i, table in enumerate(tables):
    display ("Table :",i+1)         
    df = pd.read_html(str(table))[0]
    display(df.head())  
```

```python
# Maintenant reccuperons la deuxieme table. C'est a dire celle dont l'index est 1

table = tables[1]
table_html = str(table)
df = pd.read_html(table_html)[0]
display(df)
```

```python
# Exportons la table dans un fichier CSV

df.to_csv("F:/TOP_50_entreprises.csv" ,index = False)
```
